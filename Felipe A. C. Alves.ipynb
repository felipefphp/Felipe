```python
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact
import ipywidgets as widgets
def drvn2(x1,x2,n):
    dx=(x2-x1)/(n-1)
    x= np.linspace(x1,x2,n)
    y=(np.sin(x)/x)
    dy=((((np.cos(x))*x)-(np.sin(x)))/(x**2))
    dp=[(y[i+1]-y[i])/(dx) for i in range(0,n-1)]
    dc=[(y[i+1]-y[i-1])/(2*dx) for i in range(1,n-1)]
    
    fig, axes = plt.subplots(1,2,figsize=(15, 5))
    axes[0].plot(x,dy, '-b',
                 x[0:(n-1)],dp, '-r')
    #axes[0].set_xlim([x1,x2])
    axes[0].set_title('Derivadasprogressiva em vermelho')
    
    axes[1].plot(x,dy, '-b',
                 x [1:(n-1)] ,dc,'-y')
    axes[1].set_title('Derivada de ponto central em amarelo')
    plt.show()
    #Tamanho da fígura
    plt.figure(figsize=(10, 6))
    plt.plot(x,dy, '-b')
    plt.plot(x[0:(n-1)],dp, '-r')
    plt.plot(x [1:(n-1)] ,dc,'-y', label='central')
    plt.title('Gráfico das Derivadas')
    plt.savefig('derivadas.pdf')
    plt.legend(['analítica', 'Progressiva', 'Central'], loc=2)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
interact(drvn2, x1=(-100,-10,0.1), x2=(0,100,0.1) , n=(5,200,1));
```

```python
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact
import ipywidgets as widgets

def f1(x):
    return (x**3)+(x**2)-(10*x)+10

def int_a1(x1,x2):
    return (((x2**4)/4)+((x2**3)/3)-(5*(x2**2))+(10*x2))-(((x1**4)/4)+((x1**3)/3)-(5*(x1**2))+(10*x1))

def int_n(x,y,dx):
    v_int = 0
    n=len(y)
    for i in range(0,n-1):
        v_int = v_int + ((y[i]+y[i+1])*dx)/2
    return v_int

def errof1dx(x,y,dx,x1,x2):
    erroint=np.abs(int_n(x,y,dx)-int_a1(x1,x2))
    return erroint


def graf_erro1(x1,x2,dx1,dx2,ndx):
    
    dxe=np.linspace(dx1,dx2,ndx)
    erropordx=np.zeros(ndx)
    
    for i in range(0,ndx):
        x=np.arange(x1,x2,dxe[i])
        erropordx[i]=errof1dx(x,f1(x),dxe[i],x1,x2)
    plt.plot(dxe,erropordx)
    plt.grid(True)
    plt.show()
    
interact(graf_erro1, x1=(-10, -0,0.1), x2=(0.1,10, 0.1) , dx1=0.01,dx2=0.5, ndx=(100,1000,1));
```
